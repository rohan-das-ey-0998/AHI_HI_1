```python
dataframe = df[['Feeder_Category', 'CABLE_SIZE', 'No_of_times_exceeded_60_',
       'LENGTH_OF_CABLE_MTRS_', 'NO__OF_JOINTS', 'AGE_YRS_', 'PILC', 'XLPE',
       'Average_Current', 'Peak_Current', 'Average_voltage', 'Peak_Voltage',
       'target', 'Avg_Heat_Index', 'Peak_HI', 'Avg_humidity', 'Max_Humidity',
       'Avg_TEMP', 'Max_TEMP', 'New_Summer', 'New_Monsoon', 'New_Winter']]

dataframe
```

```python
## CHecking the null values 

for i in dataframe.columns:
    print(f"{i} : {dataframe[i].isna().sum()}")
```

```python
dataframe['Feeder_Category'].fillna('RES',inplace=True)
dataframe['NO__OF_JOINTS'].fillna(dataframe['NO__OF_JOINTS'].mean(),inplace=True)
dataframe['AGE_YRS_'].fillna(dataframe['AGE_YRS_'].mean(),inplace=True)
```

```python
# %% Show general statistics
dataframe.info()
```

```python
## Feeder_Category
dataframe['Feeder_Category'] = dataframe['Feeder_Category'].map({
    'RES':0,
    'COM':1,
    'MIX':2})

dataframe['New_Summer'] = dataframe['New_Summer'].map({
                                        'N':0,
                                        'Y':1})
dataframe['New_Monsoon'] = dataframe['New_Monsoon'].map({
                                        'N':0,
                                        'Y':1})
dataframe['New_Winter'] = dataframe['New_Winter'].map({
                                        'N':0,
                                        'Y':1})

## CABLE_SIZE
dataframe['CABLE_SIZE'] = dataframe['CABLE_SIZE'].map({
    240.00 : 0,
    300.00 : 1,
    225.00 : 1,
    120.00: 2,
    185.00 : 2,
    70.00 : 2,
    0.15 : 2,
    0.20: 2,
    0.30 : 2})
```

```python
dataframe['target'] = dataframe['target'].apply(np.int64)
```

```python
dataframe.info()
```

```python
# %% Show histogram for all columns
columns = dataframe.columns
for col in columns:
    print("col: ", col)
    dataframe[col].hist()
    plt.show()
```

```python
pd.set_option('display.max_columns', None)
dataframe.head()
```

```python
from interpret.glassbox import LogisticRegression,ClassificationTree, ExplainableBoostingClassifier
from interpret import show

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
X_p = dataframe.drop(columns=['target'])
y_p = dataframe['target']
```

```python
print(X_p.shape)
print(y_p.shape)
```

```python
## splitting in train test val set

X_train, X_test, y_train, y_test = train_test_split(X_p,y_p,test_size=0.2,random_state=42)

print(f"X_train : {X_train.shape}")
print(f"X_test : {X_test.shape}")
print(f"y_train : {y_train.shape}")
print(f"y_test : {y_test.shape}")
```

## For Glass Box Model

### Logistic Regression MOdel

```python
## fit for Logistic regression model 

lr = LogisticRegression(random_state = 2023, feature_names=X_train.columns, penalty='l1',solver='liblinear')
lr.fit(X_train,y_train)
print("Training Finished")
```

```python
## Evaluting the Logistic Regression model
y_pred = lr.predict(X_test)
print(f"Accuracy : {accuracy_score(y_pred,y_test)}")
print(f"Recall : {recall_score(y_pred,y_test)}")
print(f"Precision : {precision_score(y_pred,y_test)}")
print(f"F1 : {f1_score(y_pred,y_test)}")
```

```python
# %% Explain local prediction

lr_local = lr.explain_local(X_test[:100], y_test[:100], name='Logistic Regression')
show(lr_local)
```

```python
# %% Explain global logistic regression model

lr_global = lr.explain_global(name='Logistic Regression')
show(lr_global)
```

## Decision Tree 

```python
## fit for Decision Tree model 

tree = ClassificationTree()
tree.fit(X_train,y_train)
print("Training Finished")


y_pred = tree.predict(X_test)
print(f"Accuracy : {accuracy_score(y_pred,y_test)}")
print(f"Recall : {recall_score(y_pred,y_test)}")
print(f"Precision : {precision_score(y_pred,y_test)}")
print(f"F1 : {f1_score(y_pred,y_test)}")
```

```python
# %% Explain local prediction

tree_local = tree.explain_local(X_test[:100], y_test[:100], name='Tree')
show(tree_local)
```

```python
tree_global = tree.explain_global(name='Decision Tree')
show(tree_global)
```

```python
# %% Fit Explainable Boosting Machine

ebm = ExplainableBoostingClassifier(random_state=2021)
ebm.fit(X_train, y_train) 
print("Training finished.")

y_pred = ebm.predict(X_test)
print(f"Accuracy : {accuracy_score(y_pred,y_test)}")
print(f"Recall : {recall_score(y_pred,y_test)}")
print(f"Precision : {precision_score(y_pred,y_test)}")
print(f"F1 : {f1_score(y_pred,y_test)}")
```

```python
# %% Explain local prediction
ebm_local = ebm.explain_local(X_test[:100], y_test[:100], name='Tree')
show(ebm_local)
```

```python
ebm_global = ebm.explain_global(name='Explainable Boosting Machine')
show(ebm_global)
```

### LIME ( Local Interpretable Model Agnostic Explanation )

```python
from interpret.blackbox import LimeTabular
from interpret import show

from sklearn.ensemble import RandomForestClassifier
```

```python
# Fit blackbox model

rf = RandomForestClassifier()
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)
print(f"Accuracy : {accuracy_score(y_pred,y_test)}")
print(f"Recall : {recall_score(y_pred,y_test)}")
print(f"Precision : {precision_score(y_pred,y_test)}")
print(f"F1 : {f1_score(y_pred,y_test)}")
```

```python
# Initilize Lime for Tabular data
lime = LimeTabular(model=rf.predict_proba, 
                   data=X_train, 
                   random_state=1)
lime
```

```python
# Get local explanations
lime_local = lime.explain_local(X_test[-20:], 
                                y_test[-20:], 
                                name='LIME')

show(lime_local)
```

```python
# Get local explanations
lime_local = lime.explain_local(X_test[-40:], 
                                y_test[-40:], 
                                name='LIME')

show(lime_local)
```

### SHAP( Shapley Additive Explanations )

```python
import shap
```

```python
# Fit blackbox model

rf = RandomForestClassifier()
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)
print(f"Accuracy : {accuracy_score(y_pred,y_test)}")
print(f"Recall : {recall_score(y_pred,y_test)}")
print(f"Precision : {precision_score(y_pred,y_test)}")
print(f"F1 : {f1_score(y_pred,y_test)}")
```

```python
# %% Create SHAP explainer
#explainer = shap.TreeExplainer(rf)
# Calculate shapley values for test data

# %% Create SHAP explainer
explainer = shap.TreeExplainer(rf)
# Calculate shapley values for test data
start_index = 1
end_index = 2
shap_values = explainer.shap_values(X_test[start_index:end_index])
X_test[start_index:end_index]
```

```python
# %% Investigating the values (classification problem)
# class 0 = contribution to non faulty
# class 1 = contribution to faulty
print(shap_values[0].shape)
shap_values
```

```python
# %% >> Visualize local predictions
shap.initjs()
# Force plot
prediction = rf.predict(X_test[start_index:end_index])[0]
print(f"The RF predicted: {prediction}")
shap.force_plot(explainer.expected_value[1],
                shap_values[1],
                X_test[start_index:end_index]) # for values
```

```python
## Check for another record 

# %% >> Visualize local predictions
shap.initjs()
# Force plot
prediction = rf.predict(X_test[20:21])[0]
print(f"The RF predicted: {prediction}")
shap.force_plot(explainer.expected_value[1],
                shap_values[1],
                X_test[20:21]) # for values
```

```python
# %% >> Visualize global features
# Feature summary
shap.summary_plot(shap_values, X_test)
```

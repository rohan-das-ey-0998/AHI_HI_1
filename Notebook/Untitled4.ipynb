```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

# Create a sample DataFrame with a categorical feature "Color" having three categories: "Red", "Green", "Blue"
data = pd.DataFrame({
    'Color': ['Red', 'Green', 'Red', 'Blue', 'Green', 'Blue'],
    'Target': [0, 1, 1, 0, 0, 1]
})

# Encode the categorical variable using one-hot encoding
data_encoded = pd.get_dummies(data, columns=['Color'], drop_first=True)

# Separate the target variable and the features
X = data_encoded.drop('Target', axis=1)
y = data_encoded['Target']

# Fit the logistic regression model
X = sm.add_constant(X)  # Add constant intercept
model = sm.Logit(y, X)
result = model.fit()

# Retrieve the odds ratios and their corresponding p-values
odds_ratios = pd.DataFrame(np.exp(result.params), columns=['Odds Ratio'])
p_values = pd.DataFrame(result.pvalues, columns=['p-value'])

# Print the results
print(odds_ratios)
#print(p_values)
```
